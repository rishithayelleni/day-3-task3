```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Iris.csv")

print("Dataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

df.hist(figsize=(10, 8))
plt.suptitle("Distribution of Numerical Features", fontsize=14)
plt.show()

sns.countplot(x="Species", data=df)
plt.title("Count of Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop(columns=["Species"]))
plt.title("Box Plot for Outlier Detection")
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(df.drop(columns=["Species"]).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

print("\nCorrelation Values:")
print(df.drop(columns=["Species"]).corr())

print("\nSummary of Findings:")
print("- Dataset has no missing values")
print("- Dataset contains 150 samples and 3 species")
print("- PetalLengthCm and PetalWidthCm are highly correlated")
print("- SepalWidthCm shows minor outliers")
print("- Petal features are most important for prediction")
```
